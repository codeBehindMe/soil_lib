```python
import pandas as pd
```

```python
df = pd.read_csv("training.csv")
```

```python
df
```

```python
list(filter(lambda x: x.startswith('m') != True , df.columns.values.tolist()))
```

```python
df_subset = df[['PIDN',*list(filter(lambda x: x.startswith('m'), df.columns.values))]]
```

```python
df_melt = df_subset.melt(id_vars=['PIDN'])
df_melt
```

```python
df_melt['wavelength'] = df_melt['variable'].apply(lambda x: float(x[1:]))
df_melt
```

```python
import numpy as np
```

```python
df_melt["wavelength_round"] = np.round(df_melt["wavelength"], 1)
df_melt
```

```python
df_agg = df_melt[["PIDN","wavelength_round","value"]].groupby(["PIDN","wavelength_round"]).agg("mean").reset_index()
df_agg
```

```python
df_agg["wl_str"] = df_agg["wavelength_round"].apply(lambda x: f"m{x}")
df_agg
```
